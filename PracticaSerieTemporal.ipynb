Practica Modelos Predictivos: Series Temporales aplicadas al precio de cierre del oro


En esta práctica se intentará predecir el precio de cierre del oro de los siguientes días. Para ello hemos recogido en un archivo llamado datosOro.csv los datos del oro por días, desde 2003 hasta el último trimestre de 2023. Los datos han sido extraidos directamente del FOREX, mercado de divisas mundial.


Primeramente vamos a recuperar estos datos del archivo.

```python
import pandas as pd
df = pd.read_csv('datosOro.csv')

print(df)
```

Como podemos ver, tenemos 7 columnas: 
    Date: Indica el día sobre el que se muestra la información del resto de campos. Se puede observar que el primer registro empieza el 06/05/2003, llegando los registros hasta el 25/09/2023. También hay que tener en cuenta ciertos saltos de días, debido a que en fin de semana el mercado de FOREX se cierra y no se puede operar.
    Time: Es la hora de la que se muestra el resto de información. Como vemos, toda la información se recolecto sobre un solo día a una hora concreta, aunque se podia haber hecho con otro rango de tiempos, pero nosotros queremos medir el precio de cierre del día, por lo que necesitamos solo estos datos. 
    Open: Precio de entrada del oro de dicho día medido en dólares
    High: Precio máximo del oro de dicho día medido en dólares
    Low: Precio mínimo del oro de dicho día medido en dólares
    Close: Esta será nuestra variable objetivo, ya que es lo que queremos predecir, el precio de cierre del día medido
    Volume: Número de lotes vendidos en el día, que en el caso del oro es de 100 onzas de oro por lote. (esta variable no creo que se pueda quitar por que es que es realmente la unica que afecta al precio del oro como tal)

Como hemos visto, no tiene mucho sentido quedarnos con la columna time, ya que no operamos a nivel de horas ni tiempos inferiores, por lo que esta columna no aporta valor.
